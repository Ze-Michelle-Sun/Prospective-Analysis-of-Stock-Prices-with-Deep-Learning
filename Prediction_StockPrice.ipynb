### <font color='orange'>Here we got all input variables to be used in regressor model</font>

```python
X=X_rdf.values
```

```python
# Split train and test data
X_train, X_test, y_train, y_test = train_test_split(X, yr, test_size=0.3)
```

### <font color='orange'>Creating neural network-based Regressor model </font>

```python
# Create neural network-based Regressor model 

def Regressor_model():
    
    # Creates a new sequential model using Keras
    mod=Sequential()
    
    # Adds a fully connected dense layer with 32 neurons, using the 'relu' activation function, a normal kernel initializer, and an input shape of (200,)    
    mod.add(Dense(32, kernel_initializer='normal',input_dim = 200, activation='relu'))
    
    # Adds another dense layer with 64 neurons and 'relu' activation
    mod.add(Dense(64, kernel_initializer='normal',activation='relu'))
    
    # Adds another dense layer with 128 neurons and 'relu' activation
    mod.add(Dense(128, kernel_initializer='normal',activation='relu'))
    
    # Adds another dense layer with 256 neurons and 'relu' activation
    mod.add(Dense(256, kernel_initializer='normal',activation='relu'))
    
    # Adds a final dense layer with 4 neurons and linear activation
    mod.add(Dense(4, kernel_initializer='normal',activation='linear'))
    
    # Configures the model for training with the specified loss function, optimizer, and evaluation metrics
    mod.compile(loss='mean_absolute_error', optimizer='adam', metrics=['accuracy','mean_absolute_error'])
    
    # Prints a summary of the model architecture
    mod.summary()
    
    # Returns the compiled model
    return mod
```

```python
# using the Regressor_model() function as the build function to create the neural network model
regressor = KerasRegressor(build_fn=Regressor_model, batch_size=16,epochs=2000)
```

```python
# Defining the checkpoint callback to save the best model weights during training
callback = tf.keras.callbacks.ModelCheckpoint(filepath='./Regressor_model.h5',
                                           monitor='mean_absolute_error',
                                           verbose=0,
                                           save_best_only=True,
                                           save_weights_only=False,
                                           mode='auto')
# fits the Keras regressor model
results=regressor.fit(X_train,y_train,callbacks=[callback])
```

```python
# Make predictions on the test set
y_pred= regressor.predict(X_test)
```

```python
# Creating input sequences and target variable for the training set
X_train = []
y_train = []
for i in range(50,len(train_set)):
    X_train.append(training_set_scaled[i-50:i,:])
    y_train.append(target_set_scaled[i,:])

# Converting the input sequences and target variable into NumPy arrays
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
print(X_train.shape)
```

```python
X_train.shape
```

```python
print(y_train.shape)
```

```python
# Defining the model architecture using the Sequential API
def model():
    mod = Sequential()

    # Adding the first LSTM layer with 64 units, returning sequences and specifying input shape
    mod.add(LSTM(units=64, return_sequences=True, input_shape=(X_train.shape[1], 11)))

    # Adding a dropout layer with a rate of 0.2
    mod.add(Dropout(0.2))

    # Adding a batch normalization layer to normalize the activations of the previous layer
    mod.add(BatchNormalization())

    # Adding a second LSTM layer with 64 units and returning sequences
    mod.add(LSTM(units=64, return_sequences=True))

    # Adding a dropout layer with a rate of 0.1
    mod.add(Dropout(0.1))

    # Adding a batch normalization layer to normalize the activations of the previous layer
    mod.add(BatchNormalization())

    # Adding a third LSTM layer with 64 units
    mod.add(LSTM(units=64))

    # Adding a dropout layer with a rate of 0.1
    mod.add(Dropout(0.1))

    # Adding a batch normalization layer to normalize the activations of the previous layer
    mod.add(BatchNormalization())

    # Adding a dense layer with 16 units and 'tanh' activation function
    mod.add(Dense(units=16, activation='tanh'))

    # Adding a batch normalization layer to normalize the activations of the previous layer
    mod.add(BatchNormalization())

    # Adding a dense layer with 4 units and 'tanh' activation function
    mod.add(Dense(units=4, activation='tanh'))

    # Compiling the model with 'mean_squared_error' loss function, 'adam' optimizer and metrics including accuracy and mean squared error
    mod.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy','mean_squared_error'])

    # Printing the summary of the model
    mod.summary()

    # Returning the compiled model
    return mod
```

```python
# Creating an instance of the RNN model
RNN_model=model()
```

```python
# Defining the checkpoint callback to save the best model weights during training
callback = tf.keras.callbacks.ModelCheckpoint(filepath='./RNN_model.h5',
                                              monitor='mean_squared_error',
                                              verbose=0,
                                              save_best_only=True,
                                              save_weights_only=False,
                                              mode='auto',
                                              save_freq='epoch')

# Training the RNN model on the training data with 2000 epochs and a batch size of 32, while using the checkpoint callback
RNN_model.fit(X_train, y_train, epochs=2000, batch_size=32, callbacks=[callback])
```
